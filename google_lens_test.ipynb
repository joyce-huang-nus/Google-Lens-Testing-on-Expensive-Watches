```python
# Create URL
import re
def extract_urls(string):
    regex = r"https.*png"
    match = re.findall(regex, string)
    return match
```

```python
def load_txt(path):
    with open(path) as f:
        txt_text = f.read()
    return txt_text

string = load_txt("/Users/launchpad/Downloads/kym/testset/url.txt")
url_list = extract_urls(string)
```

```python
# Run Google Lens API

import requests
#url_list = ['https://user-images.githubusercontent.com/88580416/234807651-f04e7bc5-599b-45db-877b-1e62b258b4b5.png']
URL = "https://serpapi.com/search?engine=google_lens"
def get_google_lens_result(url_list):
    model_lst = []
    for link in url_list:
        params = {
        "engine": "google_lens",
        "url": link,
        "api_key": "INSERT YOUR SERPAPI KEY"
        }
        result = requests.get(url = URL, params = params)
        data = result.json()
        isKeyPresent = 'visual_matches' in data
        if isKeyPresent == True: #it rarely happens, but sometimes the image can't get a visual match
            model = data["visual_matches"][0]['title']
        else:
            model = 'No Visual Matches'
        model_lst += [model]
    return model_lst
```

```python
import pandas as pd
model_lst = get_google_lens_result(url_list)
model_lst
```

```python
test_pdf = pd.read_csv("result.csv") # integrated into your result csv file
test_pdf['google_lens'] = model_lst
display(test_pdf.head(3))
```

# To extract watch models from the "visual match's title" 

```python
import re

def check_brands(google_lens):
    regex_01 = r"\s*(Rolex)*\s*.*Daytona"
    regex_02 = r"\s*(Rolex)*\s*.*Submariner"
    regex_03 = r"\s*(Zenith)*\s*.*El\s*.*Primero"
    regex_04 = r"\s*(Jaeger)*\s*.*(Lecoultre)*\s*.*Reverso"
    regex_05 = r"\s*(Hublot)*\s*.*Big\s*.*Bang"

    match_01 = re.search(regex_01, google_lens)
    match_02 = re.search(regex_02, google_lens)
    match_03 = re.search(regex_03, google_lens)
    match_04 = re.search(regex_04, google_lens)
    match_05 = re.search(regex_05, google_lens)
    
    if match_01 != None:
        return 'Rolex Daytona'
    elif match_02 != None:
        return 'Rolex Submariner'
    elif match_03 != None:
        return 'Zenith El Primero'
    elif match_04 != None:
        return 'Jaeger Lecoultre Reverso'
    elif match_05 != None:
        return 'Hublot Big Bang'
    else:
        return 'None'
    
```

```python
test_pdf['check'] = test_pdf['google_lens'].apply(lambda x: check_brands(x))
```
